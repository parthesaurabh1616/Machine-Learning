# Gradient Descent 2D Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function (Quadratic Function)
def loss_function(x, y):
    return (x - 5)**2 + (y - 3)**2  # Minimum at (5,3)

# Gradient of the function
def gradient(x, y):
    df_dx = 2 * (x - 5)  # Partial derivative w.r.t x
    df_dy = 2 * (y - 3)  # Partial derivative w.r.t y
    return np.array([df_dx, df_dy])

# Gradient descent parameters
learning_rate = 0.2  # Step size
max_iterations = 20   # Number of steps
tolerance = 1e-5      # Convergence tolerance

# Start position
x, y = 9, 7
path = [(x, y, loss_function(x, y))]  # Store descent points (x, y, loss)

for _ in range(max_iterations):
    grad = gradient(x, y)  # Compute gradient
    new_x, new_y = x - learning_rate * grad[0], y - learning_rate * grad[1]  # Gradient update

    # Stop if movement is very small
    if np.linalg.norm([new_x - x, new_y - y]) < tolerance:
        break

    x, y = new_x, new_y
    path.append((x, y, loss_function(x, y)))

# Convert path to numpy array for plotting
path = np.array(path)

# Create meshgrid for 3D surface plot
X = np.linspace(0, 10, 100)
Y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(X, Y)
Z = loss_function(X, Y)

# Create 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.6)

# Plot gradient descent path
ax.plot(path[:, 0], path[:, 1], path[:, 2], marker="o", color="blue", markersize=5, label="Descent Path")

# Draw arrows to indicate movement direction
for i in range(len(path) - 1):
    ax.quiver(path[i, 0], path[i, 1], path[i, 2], 
              path[i+1, 0] - path[i, 0], path[i+1, 1] - path[i, 1], path[i+1, 2] - path[i, 2], 
              color="red", alpha=0.8)

# Highlight start and end points
ax.scatter(9, 7, loss_function(9,7), color="green", s=100, label="Start (9,7)")
ax.scatter(5, 3, loss_function(5,3), color="red", s=150, marker="*", label="Local Minimum (5,3)")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Loss Function")
ax.set_title("3D Gradient Descent Path")
ax.legend()

plt.show()

# Print final position and loss value
final_x, final_y, final_loss = path[-1]
print(f"Final Position: ({final_x:.4f}, {final_y:.4f})")
print(f"Final Loss Value: {final_loss:.4f}")
```
